# 데이터 전처리

<br>

## 파일 업로드

```python
import pandas as pd

master = pd.read_csv('./csv/Master.csv')
product = pd.read_csv('./csv/Pruduct.csv')
session = pd.read_csv('./csv/Session.csv')
search1 = pd.read_csv('./csv/Search1.csv')
search2 = pd.read_csv('./csv/Search2.csv')
custom = pd.read_csv('./csv/Custom.csv')
```

<br>

## Product 테이블의 '상품브랜드(PD_BRA_NM)'명 전처리 [] 키워드 제거

```python
new = []
for x in product.PD_BRA_NM:
    new.append(x.strip('[]'))

product['PD_BRA_NM'] = new
product.head()
```

<br>

## Product(상품구매)와 Master(상품분류) 테이블 조인 (기준 : PD_C)¶

```python
print(len(product.PD_C), len(master.PD_C))
```

```python
# PD_C를 기준으로 테이블 Join
product_master = pd.merge(product, master, how='outer', on=['PD_C'])
product_master.head()
```

<br>

## Session(세션)과 Search1(검색어1) 테이블 조인 (기준 : CLINT_ID , SESS_ID)

```python
search1.head()
```

```python
# CLINT_ID와 SESS_ID을 기준으로 Join
session_search = pd.merge(session, search1, how='outer', on=['CLNT_ID', 'SESS_ID'])
session_search.head()
```

<br>

## [상품구매 + 상품분류] + [세션 + 검색어1] 테이블 조인 (기준 : CLINT_ID , SESS_ID)

```python
# CLINT_ID와 SESS_ID로 조인
pm_ss = pd.merge(product_master, session_search, how='outer', on=['CLNT_ID', 'SESS_ID'])
pm_ss.head()
```

```python
len(pm_ss)
```

<br>

## Custom(회원) + [상품구매 + 상품분류 + 세션 + 검색어1] 테이블 조인 (기준 : CLINT_ID)

```python
# CLINT_ID를 기준으로 조인

total = pd.merge(pm_ss, custom, how='outer', on=['CLNT_ID'])
total.head()
```

## total의 형태 파악

```python
len(total)
```

```python
total.shape
```

## 병합한 5개의 테이블 'total.csv'로 저장

```python
total.to_csv('./csv/total.csv', index=False)
```
